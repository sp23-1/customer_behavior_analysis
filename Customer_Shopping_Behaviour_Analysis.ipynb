```python
#To check missing values
#Mean can get affected by outliers and that would skew our distribution of our data
#Median is robust to outliers
#We will find out the median review rating within each category and replace that with missing values 
#because if we use median only then rating would be same for both clothing and footwear category 
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a columns age_group
# qcut = splits the ages into 4 equal sized-group on the basis of data distribution and assigns the label we define
#  q = quartile
# quartile = (labels) here
#Each quartile should have 12 รท 4 = 3 values.
#So pandas finds the cut points that split the sorted data into 4 equal groups.
#That means the approximate cut boundaries will be:
#Q1: Up to ~25 (25th percentile)
#Q2: Up to ~40 (50th percentile)
#Q3: Up to ~55 (75th percentile)
#Q4: Above ~55 (100th percentile)

# pd.qcut()= Quantile cut

labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annualy':365,
    'Every 3 Months':90
}
#we use map function to replace the text in the column with the coreresponding no of days
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
